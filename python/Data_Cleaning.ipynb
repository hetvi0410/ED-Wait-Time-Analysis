```python
import pandas as pd
```

```python
er = pd.read_csv("er_visits.csv")
```

```python
# 1. Clean er_visits Table
# -----------------------------------------------

# Normalize gender values
gender_map = {
    "M": "Male", "m": "Male", "MALE": "Male", "male": "Male", "Male": "Male",
    "F": "Female", "f": "Female", "FEMALE": "Female", "female": "Female", "Female": "Female",
}

def clean_gender(x):
    if pd.isna(x) or str(x).strip() in ["", " ", "unknown", "N/A"]:
        return None
    x = str(x).strip()
    return gender_map.get(x, None)

er["gender"] = er["gender"].apply(clean_gender)

# Triage cleaning
triage_map = {
    "three": 3, "Three": 3, "THREE": 3, "triage 3": 3,
    "ctas3": 3, "3 ": 3, " 3": 3, "T3": 3
}

er["triage_level"] = er["triage_level"].replace(triage_map)
er["triage_level"] = pd.to_numeric(er["triage_level"], errors="coerce")

# er["reason_for_visit"] = er["reason_for_visit"].str.replace(",", " OR ")

# Yes/No cleaning
yes_values = ["yes","Yes","YES","yEs","yeS","yess","y","1"]
no_values  = ["no","No","NO","nO","noo","nope","n","0"]

def clean_yes_no(x):
    if pd.isna(x) or str(x).strip() in ["", " ", "?", "--", "unknown", "N/A"]:
        return None
    x = str(x).strip()
    if x in yes_values:
        return "Yes"
    if x in no_values:
        return "No"
    return None

er["left_without_being_seen"] = er["left_without_being_seen"].apply(clean_yes_no)
er["revisit_within_72h"] = er["revisit_within_72h"].apply(clean_yes_no)

# Numeric conversions - Negative wait times → absolute
er["wait_time_minutes"] = pd.to_numeric(er["wait_time_minutes"], errors="coerce")
er["wait_time_minutes"] = er["wait_time_minutes"].abs().astype("Int64")

# Numeric conversions - Age to numeric
er["age"] = pd.to_numeric(er["age"], errors="coerce").astype("Int64")

# Int64 keeps age as an Integer.
# Preserves missing values as <NA>.  

# er = er.fillna("")

# Remove duplicates
er = er.drop_duplicates()

er.to_csv("er_visits_clean.csv", index=False, encoding="utf-8")
print("er_visits_clean.csv created")
```

```python
# 3. Clean physicians Table
#------------------------------------------------

physicians = pd.read_csv("physicians.csv")

physicians["physician_id"] = physicians["physician_id"].astype(str).str.strip()
physicians["name"] = physicians["name"].astype(str).str.strip()
physicians["specialty"] = physicians["specialty"].astype(str).str.strip()
physicians["years_experience"] = pd.to_numeric(physicians["years_experience"], errors="coerce")

physicians.to_csv("physicians_clean.csv", index=False)
print("physicians_clean.csv created")
```

```python
# 4. Clean nurses table
#-------------------------------------------------

nurses = pd.read_csv("nurses.csv")

nurses["nurse_id"] = nurses["nurse_id"].astype(str).str.strip()
nurses["name"] = nurses["name"].astype(str).str.strip()
nurses["years_experience"] = pd.to_numeric(nurses["years_experience"], errors="coerce")

nurses.to_csv("nurses_clean.csv", index=False)
print("nurses_clean.csv created")
```

```python
# 5. Clean beds table
#--------------------------------------------------

beds = pd.read_csv("beds.csv")

beds["bed_id"] = beds["bed_id"].astype(str).str.strip()
beds["bed_type"] = beds["bed_type"].astype(str).str.strip()
beds["department"] = beds["department"].astype(str).str.strip()

beds.to_csv("beds_clean.csv", index=False)
print("beds_clean.csv created")
```

```python
# 6. Clean er_hourly_metrics table
#--------------------------------------------------

er_hourly = pd.read_csv("er_hourly_metrics.csv")

# Datetime
er_hourly["datetime_hour"] = pd.to_datetime(er_hourly["datetime_hour"], format="%Y-%m-%d %H:%M")

# Numeric columns
er_hourly["er_occupancy"] = pd.to_numeric(er_hourly["er_occupancy"], errors="coerce")
er_hourly["staffing_level"] = pd.to_numeric(er_hourly["staffing_level"], errors="coerce")
er_hourly["avg_wait_time_last_hour"] = pd.to_numeric(er_hourly["avg_wait_time_last_hour"], errors="coerce")
er_hourly["patients_left_without_being_seen_last_hour"] = pd.to_numeric(
    er_hourly["patients_left_without_being_seen_last_hour"], errors="coerce")

er_hourly.to_csv("er_hourly_metrics_clean.csv", index=False)
print("er_hourly_metrics_clean.csv created")
```
